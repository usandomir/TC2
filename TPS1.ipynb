Tomando como norma de frecuencia a $ \omega_{0} $, escalamos los componentes del circuito para trabajar nuestra transferencia con solo un parámetro.

#### Simulación Matemática de la función transferencia normalizada en Python:


```python
from splane import pzmap, grpDelay 
from scipy import signal
import matplotlib.pyplot as plt

my_tf = signal.TransferFunction([1,-1], [1, 1])
w, mag, phase = signal.bode(my_tf)

plt.figure()
plt.ylim(-1,1)
plt.title('Magnitude response')
plt.xlabel('Angular frecuency')
plt.ylabel('Magnitude[dB]')
plt.semilogx(w, mag)    # Magnitude

plt.figure()
plt.title('Phase response')
plt.xlabel('Angular frecuency')
plt.ylabel('Phase[deg]')
plt.semilogx(w, phase)  # Phase
pzmap(my_tf)            # P&Z Map
grpDelay(my_tf)         # GroupDelay
plt.show()
plt.close('all')
```
