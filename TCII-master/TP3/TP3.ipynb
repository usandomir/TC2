# Ejercicio 1
Determinar la expresión del retardo de grupo correspondiente a las siguientes
funciones

1. $H(s)= \frac{s+1}{s^2+2s+5}$
1. $H(s)= \frac{s+2}{(s+1)(s+4)}$

Para calcular el retardo:

Siendo la ubicación de los polos

$Z_n=-\sigma_z+j\omega_z$

$P_n=-\sigma_p+j\omega_p$

    
$\tau=-\frac{d\phi}{d\omega}=-\sum \frac{\sigma_z}{\sigma_z^2+(\omega\pm \omega_z)^2}+\sum \frac{\sigma_p}{\sigma_p^2+(\omega\pm \omega_p)^2}$


Punto 1:

Por la ubicación de los polos y ceros:

$Z1=-1$

$P12=-1\pm j2$

$\tau=-\frac{1}{1+\omega^2}+\frac{1}{1+(\omega+2)^2}+\frac{1}{1+(\omega-2)^2}$

$\tau=-\frac{1}{1+\omega^2}+\frac{1}{\omega^2+4\omega+5}+\frac{1}{\omega^2-4\omega+5}$

```python
from sympy import *

w=Symbol('w') 

tau=-(1)/(1+w**2)+1/(w**2+4*w+5)+1/(w**2-4*w+5)
tau=simplify(tau)
tau
```

Otra forma:

Encontramos la fase:

$H(j\omega)=\frac{j\omega+1}{-\omega^2+j2\omega+5}$

$\phi(j\omega)=atan(\omega)-atan(\frac{2\omega}{-\omega^2+5})$

```python
fase=atan(w)-atan((2*w)/(-w**2+5))
tau=diff(-fase)
tau=simplify(tau)
tau
```

Punto 2

Por la ubicación de los polos y ceros:

$Z1=-2$

$P1=-1$

$P2=-4$

$\tau=-\frac{2}{4+\omega^2}+\frac{1}{1+\omega^2}+\frac{4}{16+\omega^2}$

```python
from sympy import *

w=Symbol('w') 

tau=-(2)/(4+w**2)+(1)/(1+w**2)+(4)/(16+w**2)
tau=simplify(tau)
tau
```

Otra forma:

Encontramos la fase:

$H(s)=\frac{s+2}{s^2+5s+4}$

$H(j\omega)=\frac{j\omega+2}{-\omega^2+j5\omega+4}$

$\phi(j\omega)=atan(\frac{\omega}{2})-atan(\frac{5\omega}{-\omega^2+4})$

```python
fase=atan(w/2)-atan((5*w)/(-w**2+4))
tau=diff(-fase)
tau=simplify(tau)
tau
```
