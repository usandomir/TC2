En donde podemos plantear las siguientes ecuaciones:

$$ V_L \cdot gm_1+V_1 \cdot gm_2+V_B \cdot gm_3=V_H \cdot gm_4 $$ 

$$ \frac{-V_H \cdot gm_5}{SC} =I_{oB} \cdot \frac{1}{SC} =V_B $$ 

$$ \frac{V_B \cdot gm_6}{SC} =I_{oL} \cdot \frac{1}{SC} =V_L $$

De la ecuación 2:

$$ V_H=-\frac{V_B \cdot SC}{gm_5} $$

De la ecuación 3

$$ V_B=\frac{V_L \cdot SC}{gm_6} $$

Reemplazando en la ecuación 1:

$$ V_L \cdot gm_1+V_1 \cdot gm_2+\frac{V_L \cdot SC}{gm_6} \cdot gm_3=-\frac{V_B \cdot SC}{gm_5} \cdot gm_4 $$ 

$$ V_L \cdot gm_1+V_1 \cdot gm_2+\frac{V_L \cdot SC}{gm_6} \cdot gm_3=-\frac{\frac{V_L \cdot SC}{gm_6} \cdot SC}{gm_5} \cdot gm_4 $$ 

Acomodando todo:

$$ V_L \cdot gm_1+V_1 \cdot gm_2+\frac{V_L \cdot gm_3 \cdot SC}{gm_6} =-\frac{V_L \cdot S^2C^2 gm_4}{gm_5 \cdot gm6} $$ 

Todo VL en un miembro y V1 del otro

$$ V_1 \cdot gm_2+ =-\frac{V_L \cdot S^2C^2 gm_4}{gm_5 \cdot gm6}-\frac{V_L \cdot gm_3 \cdot SC}{gm_6}-V_L \cdot gm_1 $$ 

Agrupando:

$$ V_1 \cdot gm_2 =-V_L \cdot \frac{S^2C^2 gm_4+gm_5 \cdot gm_3 \cdot SC+ gm_5\cdot gm_6 \cdot gm_1 }{gm_5 \cdot gm_6} $$ 

$$ \frac{V_L}{V_1}=-\frac{gm_2gm_5gm_6}{S^2C^2 gm_4+SC gm_5 gm_3 + gm_1gm_5gm_6} $$ 

$$ \frac{V_L}{V_1}=-\frac{gm_2gm_5gm_6}{C^2gm_4 \cdot(S^2+S \frac{\cancel{C} gm_5 gm_3}{C^{\cancel{2}}gm_4} + \frac{gm_1gm_5gm_6}{C^2gm_4})} $$ 

$$ \frac{V_L}{V_1}=-gm_2 \cdot \frac{\frac{gm_5gm_6}{C^2 gm_4}}{(S^2+S \frac{ gm_5 gm_3}{ C gm_4} + \frac{gm_1gm_5gm_6}{C^2gm_4})} $$ 


$$ \frac{V_L}{V_1}=-\frac{gm_2}{gm_1} \cdot \frac{\frac{gm_1gm_5gm_6}{C^2gm_4}}{(S^2+S \frac{ gm_5 gm_3}{ C gm_4} + \frac{gm_1gm_5gm_6}{C^2gm_4})} $$ 

En donde:

$$ K=-\frac{gm_2}{gm_1} $$

$$ \omega_n^2= \frac{gm_1gm_5gm_6}{C^2 gm_4} $$

$$ \frac{\omega_n^2}{Q^2}=\frac{ gm_5^2 gm_3^2}{ C^2 gm_4^2} $$

$$ Q^2 = \frac{\cancel{C^2} gm_4^{\cancel{2}}}{gm_5^{\cancel{2}} gm_3^2} \cdot \frac{gm_1 \cancel{gm_5} gm_6}{\cancel{C^2} \cancel{gm_4}} $$ 

Ecuaciones de diseño:

$$ K=-\frac{gm_2}{gm_1} $$

$$ Q= \frac{1}{gm_3} \cdot \sqrt{\frac{gm_1 gm_4 gm_6}{gm_5}}  $$

$$ \omega_n= \frac{1}{C} \cdot \sqrt{\frac{gm_1gm_5gm_6}{gm_4}} $$

Si $gm_1=_gm_4=gm_5=gm_6=gm $

$$ K=-\frac{gm_2}{gm} $$

$$ Q= \frac{gm}{gm_3}  $$

$$ \omega_n= \frac{gm}{C} $$

En donde con:

gm_2 modificamos la ganancia

gm_3 el Q

C la frecuencia de corte \frecuencia central (pasa banda)

Ej:

Filtro pasabajo a 10KHz Q= 0,707 (Butterworth) Ganancia en continua=-1

Proponemos:gm=1ms

Obtenemos:

$$ gm_2=1ms $$

$$ gm_3= \sqrt{2}ms$$

$$ C=15,91nf$$

### Modificando la estructura


Si vemos la estructura planteada, podemos sumar directamente en el OTA5, quedando la transconductancia en paralelo con la capacidad, lo que hace que queda anulada la contribución del OTA3. por lo tanto el circuito nos queda:

Vamos a encontrar las expresiones que nos quedan ahora:

$$ V_L \cdot gm_1+V_1 \cdot gm_2=-V_B(gm_5+SC)$$ 

$$ \frac{V_B \cdot gm_6}{SC} =I_{oL} \cdot \frac{1}{SC} =V_L $$

Reemplazando:

$$ V_L \cdot gm_1 + V_1 gm_2=-V_L \frac{SC}{gm_6} \cdot (gm_5+SC) $$

$$ V_L \cdot gm_1 + V_1 gm_2=-V_L (\frac{S^2C^2}{gm_6}+SC \frac{gm5}{gm6} + gm1) $$

$$ \frac{V_L}{V1}=\frac{gm_2}{\frac{C^2}{gm_6}(S^2+S \frac{gm_5}{C}+\frac{gm_1 gm_6}{C^2})}$$

$$ \frac{V_L}{V1}=-\frac{gm_2}{gm_1} \cdot \frac{\frac{gm_1 gm6}{C^2}}{(S^2+S \frac{gm_5}{C}+\frac{gm_1 gm_6}{C^2})}$$

$$ \omega_n^2= \frac{gm_1gm_6}{C^2} $$

$$ \frac{\omega_n}{Q}=\frac{gm_5}{C} $$

$$ Q=\frac{\sqrt{gm_1 gm_6}}{\cancel{C} \frac{gm_5}{\cancel{C}}} $$

$$ Q=\frac{\sqrt{gm_1 gm_6}}{gm_5} $$

$$ K=-\frac{gm_2}{gm_1} $$

si $gm_1=gm_6=gm$

$$ K=-\frac{gm_2}{gm} $$

$$ \omega_n= \frac{gm}{C} $$

$$ Q=\frac{gm}{gm_5} $$

Para el ejemplo anterior:
