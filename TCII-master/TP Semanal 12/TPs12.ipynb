Vamos a sintetizar Y11 cumpliendo los ceros de la transferencia presentes en Y21, para eso proponemos un polonomio denominador en ambas expresiones a fin que Y11 sea FRP y posea alternancia de polos y ceros quedano:

Sintesis Analítica:

$Y_{11}=\frac{(s+2)(s+4)}{s+3}$

$Y_2=Y_{11}-K_1$

$K_1=\lim_{s \to -1}(Y_{11})=\frac32$

$Y_2=Y_{11}-\frac32=\frac{s^2+6s+8}{s+3}-\frac32=\frac{2s^2+12s+16-3s-9}{2(s+3)}=\frac{2s^2+9s+7}{s(s+3)}$

$Y_2=\frac{(s+1)(s+\frac72)}{s+3}$

$Z_2=\frac{s+3}{(s+1)(s+\frac72)}$

$Z_4=Z_2-\frac{K_2}{s+1}$

$K_2=\lim_{s \to -1}(Z_2 (s+1)=\lim_{s to -1}(\frac{s+3}{s+\frac72})$

$K_2=\frac45$

La impedancia de un capacitor en paralelo con un resistor vale:

$\frac{\frac1c}{s+\frac{1}{RC}}=\frac{\frac45}{s+1}$

De ahi que el valor de $C=\frac54$ y el valor de $R=\frac45$

$Z_4=\frac{s+3}{(s+1)(s+\frac72)}-\frac{4}{5(s+1)}=\frac{5s+15-4s-14}{5(s+1)(s+\frac72)}=\frac{1}{5(s+\frac72)}$

$Y_4=5(s+\frac72)$

$Y_6=Y_4-K_3 \cdot s$

$K_3=\lim_{s \to \infty}(\frac{5(s+\frac72)}{s})=5$

$Y_6=5s-\frac{35}{2}-5s=\frac{35}{2}$

$Z_6=\frac{2}{35}$
