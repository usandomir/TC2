$ \frac{V2(s)}{V1(s)}= \frac{1-SCR_3}{1+SCR_3} $

$ \frac{V2(j\omega)}{V1(j\omega)}= \frac{1-j\omega CR_3}{1+j\omega CR_3} $

$ \left | \frac{V2(j\omega)}{V1(j\omega )} \right |= \frac{\sqrt{1+(-j\omega CR_3)^{2}}}{\sqrt{1+(j\omega CR_3)^{2}}} $  
Si normalizamos para $ \omega =1000 \frac{rad}{seg} $ y para $ R=1000\Omega  $ 

$ \left | \frac{V2(j\omega)}{V1(j\omega )} \right |= \frac{\sqrt{1+(-j\omega_n )^{2}}}{\sqrt{1+(j\omega_n )^{2}}}=1  \forall  \omega \in \mathbb{R} $ 

$ \theta (j\omega)=arctg(-\omega_n)-arctg(\omega_n)  $

Como  $ arctg(x) $ es una funcion impar  $ arctg(x)=-arctg(-x)  $ 

$ \theta (j\omega)=-arctg(\omega_n)-arctg(\omega_n)  $

$ \theta (j\omega)=-2arctg(\omega_n) $






```python
from scipy import signal
import matplotlib.pyplot as plt
from splane import bodePlot, pzmap
import numpy as np

R=1000 #ohms
C=10**(-6) #faradios
Rn=R/1000 
wn=1000
Cn=(C*wn*R)
start_omega = 10**-2 # r/s
stop_omega = 10**2 # r/s
cant_puntos= 10000 # puntos
omega=np.linspace( start_omega, stop_omega , cant_puntos)
sys = signal.TransferFunction([-Cn*Rn,1], [Cn*Rn,1])
w, mag, phase = signal.bode(sys,omega)

plt.figure()
plt.semilogx(w, mag)    # Bode magnitude plot
plt.ylim([-1,1])

plt.figure() 
plt.semilogx(w, phase)  # Bode phase plot
pzmap(sys)

plt.show()
```
