## Estudio de Sensibilidad

\begin{equation} 
    \omega_n= \frac{1}{CR_3}  
\end{equation}
\begin{equation} 
    S_{C}^{\omega_n} =\frac{C}{\omega_n} \cdot \frac{\partial{\omega_n}}{\partial{C}}=\frac{\cancel{C}}{\omega_n} \cdot -\frac{1}{R_3 \cancel{C^2}}
\end{equation}
\begin{equation} 
    S_{C}^{\omega_n} = -\frac{1}{\cancel{\omega_n}} \cdot \cancel{\omega_n} = -1
\end{equation}
    
\begin{equation} 
    S_{Q}^{R_2}
\end{equation}

\begin{equation} 
    Q=\frac{R_2}{R_3}
\end{equation}

\begin{equation} 
   S_{Q}^{R_2} =\frac{R_2}{Q} \cdot \frac{\partial{Q}}{\partial{R_2}}
\end{equation}

\begin{equation} 
   S_{Q}^{R_2} = \frac{R_2}{Q} \cdot \frac{1}{R_3} = \frac{\cancel{R_2}}{\cancel{R_3}} \cdot \frac{\cancel{R_3}}{\cancel{R_2}}=1
\end{equation}

\begin{equation} 
   S_{Q}^{R_3} =\frac{R_3}{Q} \cdot \frac{\partial{Q}}{\partial{R_3}}
\end{equation}

\begin{equation} 
   S_{Q}^{R_3} = \frac{\cancel{R_3}}{Q} \cdot -\frac{R_2}{R_3^{\cancel{2}}} = -\frac{\cancel{R_3}}{\cancel{R_2}} \cdot \frac{\cancel{R_2}}{\cancel{R_3}}=-1
\end{equation}
