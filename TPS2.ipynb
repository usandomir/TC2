Si conocemos el orden de nuestro sistema, $n = 2$, sabemos también que nuestro único par de polos conjugados debe estar a $\frac{\pi}{4}\frac{rad}{s}$ del eje $\sigma$ del mapa de polos y ceros.

Existen 2 nodos de interés en el circuito, uno es el de $V_B$, facilmente analizable ya que es un amplificador inversor de ganancia unitaria donde se cumple la relación $V_B = -V_o$.
El otro noto de interés es el de $V_A$, que para expresar la transferencia de tensión entre ese nodo y la señal de entrada del filtro, basta con reemplazar en la función transferencia $T_{(s)} = \frac{V_o}{V_i}$ a $V_o$ en función de $V_A$ según la relación anteriormente deducida.
<br><br>
$V_o = -V_A \frac{1}{sCR_3}$
<br>
<br>Obteniendo una transferencia de Filtro Pasa-Banda:
<br><br>
$\frac{V_A}{V_i} = -\frac{R_3}{R_1} . \frac{s\omega_0}{s^2 + s \frac{\omega_0}{Q} + \omega_0^2}$
